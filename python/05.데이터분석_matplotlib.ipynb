### Matplotlib

* 시각화의 대표적인 Library로서 파이썬에서 Chart나 Plot으로 시가화(Visualization)하는 패키지이다.
* Matplotlib은 차트나 플롯이외에도 저수준의 API를 사용한 다양한 시각화기능을 제공한다.
 - Line Plot
 - Scatter Plot
 - Bar Chart
 - Histogram
 - Box Plot
 
* Maplotlib패키지에는 서브패키지인 `pylab`이 존재하는데 이 패키지는 matlab이라는 수치해석 소프트웨어의 시각화명령을 그대로 사용할 수 있도록 Matplotlib의 하위명령들을 Wrapping한 명령어 집합을 제공한다. 간단한 시각화 프로그램을 만드는 경우에는 pylab서브 패키지 명령만으로도 충분하다.

* 참고사이트 : https://matplotlib.org/tutorials/index.html

* 매직명려어(%)중 하나인 `%matplotlib`명령은 그래프를 조잘할 수 있는 명령으로서 여러가지 옵션이 있는데 아래와 같은 옵션을 많이 사용한다.
 - `inline` : 최종 그래프를 웹브라우저에 출력(그래프가 완성된 형태로 사용조작이 불가능)
 - `nbagg`  : 이 옵션 그래프를 한 단계씩 고쳐가면서 그래프를 조작할 수 있는 옵션

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 한글처리
from matplotlib import rc, font_manager
font_name = font_manager.FontProperties(fname='c:/windows/fonts/malgun.ttf')\
                .get_name()
rc('font', family=font_name)
rc('axes', unicode_minus=False) # -표시를 안깨지도록 하는 옵션
```

###### 1) 신장과 체중

```python
height = [165,177,160,180,185,155,172]
weight = [62,67,55,74,90,43,64]
print(len(height), len(weight))
```

```python
plt.scatter(height, weight)
plt.title('신장과 체중의 상관관계')
plt.xlabel('신장(cm)')
plt.ylabel('체중(kg)')
plt.grid()
plt.show()
```

###### 2) 인구밀도(2017년기준)

```python
cities = ['서울','인천','대전','대구','울산','부산','광주',]

# 도시별 위치(위/경도)
lat = [37.56, 37.45, 36.36, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

# 2017년도 통계청자료 인구밀도(명/km^2)
pop = [16154, 2751, 2839, 2790, 1099, 4454, 2995]
```

```python
size = np.array(pop) * 0.2
colors = ['r','g','b','c','m','k','y',]

plt.scatter(lon, lat, c=colors, s=size)
plt.title('2017년 지역별 인구밀도')
plt.xlabel('경도')
plt.ylabel('위도')
plt.grid()

for x,y,city in zip(lon, lat, cities):
    plt.text(x, y, city)

plt.show()
```
