```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib

%whos
```

```python
# 1. 난수발생
data = np.random.rand() 
print(data)

data = np.random.rand(50) 
print(data)
print(type(data))

data = np.random.rand(3,2,2,4) 
print(data)


#?np.random.rand
```

```python
# 2. pandas
# 1) 1차원자료구조 Series
data = np.random.rand(50)
s = pd.Series(data)
print(type(s), s)
print(dir(s))
```

```python
# 시각화(pandas고유기능으로 그래프그리기)
#?s.plot
s.plot()
```

```python
# 2) 2차원자료구조 Pandas
# 2차원자료를 생성 : np.ndarray()
# ?np.ndarray
# numpy로 랜덤하게 10행 3열의 2차원배열(정확하게는 행렬이리고 한다.)
data = np.random.rand(10,3)
data
```

```python
# pd.DataFrame?
df = pd.DataFrame(data)
display(df)
df = pd.DataFrame(data, columns=['2018', '2019', '2020'])
df
```

```python
# 시각화 : 2차원데이터로 자동그리기
df.plot()
```

```python
df.plot(kind='area')
```

```python
# pichart
ser = pd.Series(data=np.random.rand(5), 
                index=['2015', '2016', '2017', '2018', '2019'],
                name="Sales Amount")

ser.plot(kind='pie', figsize=(10,5), fontsize=15, autopct='%.2f %%')
```
